```python
import pandas as pd
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Dave'],
    'age': [25, 35, 40, 28],
    'gender': ['F', 'M', 'M', 'M'],
    'salary': [50000, 70000, 60000, 80000]
}

df = pd.DataFrame(data)

age_filtered = df[df['age'] > 30]
print("Rows where age is greater than 30:")
print(age_filtered)

name_filtered = df[df['name'].str.contains('e', case=False, na=False)]
print("\nRows where name contains 'e':")
print(name_filtered)

gender_salary_filtered = df[(df['gender'] == 'M') & (df['salary'] > 65000)]
print("\nRows where gender is 'M' and salary is greater than 65000:")
print(gender_salary_filtered)

# d. Select columns 'name' and 'age'
selected_columns = df[['name', 'age']]
print("\nSelected columns 'name' and 'age':")
print(selected_columns)
```
